```python
import sqlite3
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
db_path = "./db.sqlite3"

connection = sqlite3.connect(db_path)
```

```python
database_data = pd.read_sql_query(
    """
        SELECT * 
        FROM sqlite_master;
    """,
    connection
)
```

```python
database_data
```

```python
table_names = database_data[database_data["type"] == "table"]["name"]
table_names
```

```python
tables_dict = dict()

for table_name in table_names:
    table = pd.read_sql(
        f"SELECT * FROM {table_name}", connection
    )
    tables_dict[table_name] = table
```
